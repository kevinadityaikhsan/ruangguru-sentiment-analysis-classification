```python
from youtube_comment_downloader import *
```

```python
downloader = YoutubeCommentDownloader()
```

# Episode 1

```python
comments = downloader.get_comments_from_url("https://www.youtube.com/watch?v=dY_3acmFFRQ")

comments_list = []

for comment in comments:
  comments_list.append(comment['text'])
```

# Episode 2

```python
comments = downloader.get_comments_from_url("https://www.youtube.com/watch?v=n2sggHMSAro")

for comment in comments:
  comments_list.append(comment['text'])
```

# Episode 3

```python
comments = downloader.get_comments_from_url("https://www.youtube.com/watch?v=IFG-F5HJDpg")

for comment in comments:
  comments_list.append(comment['text'])
```

```python
import csv

# Save to CSV
with open('ruangguru_eps1-3.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['comment'])
    for comment in comments_list:
        writer.writerow([comment])
```
